```python
#
df['Location_split'] = df['Location'].str.split(',')

#
df['City'] = df['Location_split'].apply(lambda x: x[0].strip() if len(x) > 0 else None)
df['State'] = df['Location_split'].apply(lambda x: x[1].strip() if len(x) > 1 else None)

#
df.drop('Location_split', axis=1, inplace=True)
df.drop('Location', axis=1, inplace=True)
df.head()
```

```python
col_list = list(df.columns)
position = col_list.index('Company Name')
col_list.insert(position + 1, 'City')
col_list.insert(position + 2, 'State')
df = df[col_list]
df = df.loc[:, ~df.columns.duplicated()]

df.head()
```

```python


df['Headquarters_split'] = df['Headquarters'].str.split(',')
df['Headquarters_city'] = df['Headquarters_split'].apply(lambda x: x[0].strip() if len(x) > 0 else None)
df['Headquarters_state'] = df['Headquarters_split'].apply(lambda x: x[1].strip() if len(x) > 1 else None)

#
df.drop('Headquarters_split', axis=1, inplace=True)
df.drop('Headquarters', axis=1, inplace=True)

col_list = list(df.columns)
position = col_list.index('State')
col_list.insert(position + 1, 'Headquarters_city')
col_list.insert(position + 2, 'Headquarters_state')

df = df[col_list]
df = df.loc[:, ~df.columns.duplicated()]

df.head()
```

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'

def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
```

```python
df['job_simplification'] = df['Job Title'].apply(title_simplifier)

df.job_simplification.value_counts()
```

```python
df['Seniority'] = df['Job Title'].apply(seniority)
df.seniority.value_counts()
```
